### Similar tags to tags applied

```python
user_tag_app = user_tag_app[user_tag_app['tag'].isin(tag_candidates)]
```

```python
# Find similar words for tags applied
similar_words = []
for tag in user_tag_app['tag']:
    if tag in word_vec_model1.wv.vocab:
        for (word, similarity) in word_vec_model1.wv.similar_by_word(tag):
            similar_words.append(word)
            
            
# Find similar words for tags from tag searched

# Find similar words for tags from movie rated
for tag in user_tag_preference_small['tag']:
    if tag in word_vec_model1.wv.vocab:
        for (word, similarity) in word_vec_model1.wv.similar_by_word(tag):
            similar_words.append(word)
            
```

```python
# Find similar tags among similar words for tags applied
similar_tags = []

for word in similar_words:
    if word in tag_candidates:
        similar_tags.append(word)
```

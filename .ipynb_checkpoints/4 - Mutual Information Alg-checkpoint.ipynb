### Pu,t(l): Proportion of Tags applied to movies in bucket l

Pu,t(l): probability that tag t has the bucketed relevance l for all movies that user u rated

Denominator: # tag t applied to movies that the user rated

### Pu,t(r, l)
TODO:
Pu,t(r, l) is the joint-probability of the two random variables, indicating the probability that user u gives a rating r for all movies that have tag t with bucketed relevance l

Denom: # movies the user rated with tag t 

```python
tag_relevance_rating_agg = pd.merge(movie_tag_applied_relevance, user_rating_pairs, on="movieId",how="left")
tag_relevance_rating_agg
```

```python
tag_relevance_rating_agg2 = tag_relevance_rating_agg.groupby(["tag","relevance_bucket","rating"]).count().reset_index()
tag_relevance_rating_agg2.drop(["movieId","score","relevance","userId","tstamp","source"],axis=1,inplace=True)
tag_relevance_rating_agg2.rename(columns={"numApps":"count_tag_relevance_rating"},inplace=True)

```

```python
tag_relevance_agg = tag_relevance_rating_agg.groupby(["tag","relevance_bucket"]).count().reset_index()
tag_relevance_agg.drop(["movieId","score","relevance","userId","tstamp","source","rating"],axis=1,inplace=True)
tag_relevance_agg.rename(columns={"numApps":"count_tag_relevance"},inplace=True)

```

```python
tag_relevance_rating_count_agg = pd.merge(tag_relevance_rating_agg2, tag_relevance_agg, on=["tag","relevance_bucket"],how="left")
#tag_relevance_rating_count_agg["Pu,t(r,l)"] = tag_relevance_rating_count_agg['numApps_x']/tag_relevance_rating_count_agg['numApps_y']
#tag_relevance_rating_count_agg.sort_values(by=["tag","rating"],inplace=True)
tag_relevance_rating_count_agg["Putrl"] = tag_relevance_rating_count_agg["count_tag_relevance_rating"]/tag_relevance_rating_count_agg["count_tag_relevance"]
tag_relevance_rating_count_agg
```

### Calculate score and make predictions

```python
import math 

test = tag_relevance_rating_count_agg
sums=[]

for index, row in test.iterrows():
    tag = row[0]
    #print(tag)
    r = row[2]
    #print(r)
    l = row[1]
    #print(l)
    Pur = probs_a_rating[r]
    Putl = movie_tag_applied_count_agg.loc[(movie_tag_applied_count_agg['tag']==tag) & (movie_tag_applied_count_agg['relevance_bucket']==l), 'Pu,t(l)'].iloc[0]
    Putrl = row[5]
    #print(Pur)
    #print(Putl)
    #print(Putrl)
    sum = Putrl * math.log(Putrl/(Putrl*Pur),2)
    sums.append(sum)
    #print(sum)
    
```

```python
value = movie_tag_applied_count_agg.loc[(movie_tag_applied_count_agg['tag']=='action') & (movie_tag_applied_count_agg['relevance_bucket']==7), 'Pu,t(l)']
```

```python
test["MI_score"] = sums

test.sort_values(by="MI_score",ascending=False).head(20)
```

### Normalize MI score

```python
#HRut 
tag_relevance_rating_agg3 = tag_relevance_rating_agg.groupby(["tag","rating"]).count().reset_index()
tag_relevance_rating_agg3.drop(["movieId","score","relevance","relevance_bucket","userId","tstamp","source"],axis=1,inplace=True)
tag_relevance_rating_agg3.rename(columns={"numApps":"count_tag_rating"},inplace=True)
```

```python
#HRut 
tag_relevance_rating_agg4 = tag_relevance_rating_agg.groupby(["tag"]).count().reset_index()
tag_relevance_rating_agg4.drop(["movieId","score","relevance","relevance_bucket","userId","tstamp","source","rating"],axis=1,inplace=True)
tag_relevance_rating_agg4.rename(columns={"numApps":"count_tag"},inplace=True)
```

```python
tag_relevance_rating_HRut = pd.merge(tag_relevance_rating_agg3,tag_relevance_rating_agg4,on="tag",how="left")
tag_relevance_rating_HRut["Put"] = tag_relevance_rating_HRut["count_tag_rating"]/tag_relevance_rating_HRut["count_tag"]
```

```python
tag_relevance_rating_HRut["HRut"] = -tag_relevance_rating_HRut["Put"]*np.log2(tag_relevance_rating_HRut["Put"])
tag_relevance_rating_HRut_sum = tag_relevance_rating_HRut.groupby("tag").sum().reset_index()
```

```python
tag_relevance_rating_HRut_sum.sort_values(by="HRut",ascending=False)
```
